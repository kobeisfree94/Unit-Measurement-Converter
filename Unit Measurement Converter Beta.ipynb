```python
def start():
    print("Welcome to the Unit Conversion Program")

    intro= (input("Would You Like To Continue? 'Yes' or 'No':"))
    if intro == 'yes':
        print_menu()
    else:
        the_end()
```

```python
def print_menu():
    print("\nMENU")
    print("\nDISTANCE:")
    print(' 1.Meters to Inches')
    print(' 2.Inches to Meters')
    print("\nTEMPERATURE:")
    print(' 3.Farenheit to Celsius')
    print(' 4.Celsius to Farenheit')
    print("\nWeight:")
    print(' 5.Grams to Ounces')
    print(' 6.Ounces to Grams')
    print("\nVOLUME:")
    print(' 7.Liters to Cups')
    print(' 8.Cups to Liters')

    choice()
```

```python
def choice():
    choice= input("Which Conversion Would You Like To Make?: ")

    if choice == "1":
        m_inch()
    if choice == "2":
        inch_m()
    if choice == "3":
        f_c()
    if choice == "4":
        c_f()
    if choice == "5":
        g_oz()
    if choice == "6":
        oz_g()
    if choice == "7":
        ml_cup()
    if choice == "8":
        cup_ml()
```

```python
def the_end():
    print ("\nGOOD BYE!")
```

```python
def m_inch():
    m = float(input("\nEnter a Distance in Meters: "))
    inch = m*39.37 
    
    
    print("\nYou entered {0}. Distance in Miles: {1}".format(m,round(inch/63360,2)))
   
    print("\nYou entered {0}. Distance in Yard: {1}".format(m,round(inch/36,2)))
    
    print("\nYou entered {0}. Distance in Feet: {1}".format(m,round(inch/12,2)))
    
    print("\nYou entered {0}. Distance in Inches: {1}".format(m,round(inch,2))) 
    
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
        
 
```

```python
def inch_m():
    inch = float(input("\nEnter a distance in Inches: "))
    m = inch/39.37
    
    
    print("\nYou entered {0}. Distance in Killometers: {1}".format(inch,round(m/1000,2)))
    
    print("\nYou entered {0}. Distance in Meters: {1}".format(inch,round(m,2)))
    
    print("\nYou entered {0}. Distance in Centimeters: {1}".format(inch,round(m*100,2)))
    
    print("\nYou entered {0}. Distance in Millimeters: {1}".format(inch,round(m*1000,2)))
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def f_c():
    f = float(input("\nEnter a Temperature in Farenheit: "))
    c = (f-32)*(5/9)
    
    
    print("\nYou entered {0}. Temperature in Celcius: {1}".format(f,round(c,2)))
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def c_f():
    c = float(input("\nEnter a Temperature in Celcius: "))
    f = c/5*9+32
    
    
    print("\nYou entered {0}. Temperature in Farenheit: {1}".format(c,round(f,2)))
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def g_oz():
    g = float(input("\nEnter a Weight in Grams: "))
    oz = g/28.35
    
    
    print("\nYou entered {0}. Weight in Pounds: {1}".format(g,(round(oz/16,2))))
    
    print("\nYou entered {0}. Weight in Ounces: {1}".format(g,oz))
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def oz_g():
    oz= float(input("\nEnter a Weight in Ounces: "))
    g= oz*28.35
    
    
    print("\nYou entered {0}. Weight in Killograms: {1}".format(oz,(round(g/1000,2))))
    
    print("\nYou entered {0}. Weight in Grams: {1}".format(oz,round(g,2)))
    
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def ml_cup():
    ml= float(input("\nEnter a Volume in Milliters: "))
    cup= ml/237
    
    
    print("\nYou entered {0}. Volume in Teaspoons: {1}". format(ml,round(cup*48)))
    
    print("\nYou entered {0}. Volume in Tablespoons: {1}". format(ml,round(cup*16)))
    
    print("\nYou entered {0}. Volume in Cups: {1}". format(ml,round(cup,2)))
    
    print("\nYou entered {0}. Volume in Quarts: {1}". format(ml,round(cup/4)))
    
    print("\nYou entered {0}. Volume in Gallons: {1}". format(ml,round(cup/16)))

    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```

```python
def cup_ml():
    cup= float(input("\nEnter a volume in Cups: "))
    ml= cup/0.0042267528
    
    
    print("\nYou entered {0}. Volume in Liters: {1}".format(cup,round(ml/1000,2)))
    
    print("\nYou entered {0}. Volume in Milliters: {1}".format(cup,round(ml,2)))
    
    
    choice2= input("\nWould You Like to Make Another Conversion? 'Yes' or 'No': ")
    if choice2== 'yes':
        print_menu()
        choice()
    elif choice2== 'no':
        the_end()
    else:
        choice2
```
